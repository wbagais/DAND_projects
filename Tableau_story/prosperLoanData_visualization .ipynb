# Design: 

### 1st dashboard:
- Bar to represent the relationship of the employment status with stated monthly income and average debt to income ration. Then I added "is borrower homeowner" as a colour.
- (in the 1st version only) Heat map to represent the relationship between the employment status and the listing category. the size is the Debt To Income Ratio and the colour is Stated Monthly Income 
- (in the second version only) the relationship of the listed category with stated monthly income and average debt to income ration. Then I added "is borrower homeowner" as a colour. 
- Map to represent the Debt To Income Ratio in each state 

### 2nd dashboard:
- All graphs in the 1st dashboard coloured by the term.
- Bubble to represent the number of loans in each category. bubble shows the difference in size in a simple way. 
- Treemap to represent the relationship between the loan status(the text) and the number of records (the size). 
- Line chart to represent the relation between the listing creation date per month and the average loan original amount.

# Data Cleaning:

```python
import pandas as pd
import numpy as np

df = pd.read_csv('prosperLoanDataCopy.csv')
df.head()
```

```python
df.info()
```

```python
df_clean = df.copy()
```

```python
df_clean.rename(index=str, columns={"ListingCategory (numeric)": "ListingCategory"}, inplace = True)
```

```python
df_clean.info()
```

#ListingCategory	The category of the listing that the borrower selected when posting their listing: 
- 0 - Not Available, 1 - Debt Consolidation, 2 - Home Improvement, 3 - Business, 4 - Personal Loan, 5 - Student Use, - 6 - Auto,7- Other, 8 - Baby&Adoption, 9 - Boat, 10 - Cosmetic Procedure, 
- 11 - Engagement Ring, 12 - Green Loans, 13 - Household Expenses, 14 - Large Purchases, 15 - Medical/Dental, 
- 16 - Motorcycle, 17 - RV, 18 - Taxes, 19 - Vacation, 20 - Wedding Loans

```python
df_clean['ListingCategory'].replace(0, np.nan, inplace = True)
df_clean['ListingCategory'].replace(1, 'Debt_Consolidation', inplace = True)
df_clean['ListingCategory'].replace(2, 'Home_Improvement', inplace = True)
df_clean['ListingCategory'].replace(3, 'Business', inplace = True)
df_clean['ListingCategory'].replace(4, 'Personal_Loan', inplace = True)
df_clean['ListingCategory'].replace(5, 'Student_Use', inplace = True)
df_clean['ListingCategory'].replace(6, 'Auto', inplace = True)
df_clean['ListingCategory'].replace(7, 'Other', inplace = True)
df_clean['ListingCategory'].replace(8, 'Baby&Adoption', inplace = True)
df_clean['ListingCategory'].replace(9, 'Boat', inplace = True)
df_clean['ListingCategory'].replace(10, 'Cosmetic_Procedure', inplace = True)
df_clean['ListingCategory'].replace(11, 'Engagement_Ring', inplace = True)
df_clean['ListingCategory'].replace(12, 'Green_Loans', inplace = True)
df_clean['ListingCategory'].replace(13, 'Household_Expenses', inplace = True)
df_clean['ListingCategory'].replace(14, 'Large_Purchases', inplace = True)
df_clean['ListingCategory'].replace(15, 'Medical/Dental', inplace = True)
df_clean['ListingCategory'].replace(16, 'Motorcycle', inplace = True)
df_clean['ListingCategory'].replace(17, 'RV', inplace = True)
df_clean['ListingCategory'].replace(18, 'Taxes', inplace = True)
df_clean['ListingCategory'].replace(19, 'Vacation', inplace = True)
df_clean['ListingCategory'].replace(20, 'Wedding_Loans', inplace = True)
```
